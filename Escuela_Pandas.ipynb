**Pandas**

```python
# Lee un archivo JSON desde una URL y almacena los datos en un DataFrame llamado 'df'
df = pd.read_json("https://raw.githubusercontent.com/ElProfeAlejo/tratamiento_datos_python/main/ranking.json")

# Filtra el DataFrame 'df' para seleccionar solo las filas donde la columna 'sexo' es igual a 'M' (sexo masculino),
# luego toma una muestra aleatoria de 200 de estas filas y almacénalas en el DataFrame 'nombres_m'
nombres_m = df[df['sexo'] == 'M'].sample(200)

# Filtra el DataFrame 'df' para seleccionar solo las filas donde la columna 'sexo' es igual a 'F' (sexo femenino),
# luego toma una muestra aleatoria de 200 de estas filas y almacénalas en el DataFrame 'nombres_f'
nombres_f = df[df['sexo'] == 'F'].sample(200)

# Imprime el DataFrame 'nombres_f', que contiene los nombres de las mujeres muestreados aleatoriamente
nombres_f
```

# **3. Relacion**

Vamos a establecer una relación entre nuestros dataframes. En el dataframe **nombres**, utilizaremos la columna **matrícula** para relacionarla con el dataframe **cursos** a través de la columna **id_curso**.

```python
nombres.head()
```

```python
cursos.head()
```

* np.random.rad()

```python
# Generar valores aleatorios utilizando np.random.rand
x = np.random.rand(len(cursos))

# Calcular las probabilidades dividiendo cada valor aleatorio por la suma de todos los valores aleatorios
prob = x / sum(x)

# Imprimir las probabilidades
prob
```


1. Se utiliza **np.random.rand(len(cursos))** para generar valores aleatorios en función de la cantidad de cursos. Esto crea un arreglo de números aleatorios entre 0 y 1.

2. Luego, se calculan las probabilidades dividiendo cada valor aleatorio por la suma de todos los valores aleatorios. Esto normaliza los valores y asigna probabilidades a cada identificador de curso en función de estos valores aleatorios.

* iterrws()

```python
# Crear una lista llamada 'todas_matriculas' que contendrá las relaciones entre id_cursos y matrículas
todas_matriculas = []

# Iterar a través de las filas del DataFrame 'nombres' utilizando iterrows()
for index, row in nombres.iterrows():
    id_curso = row.id_alumno
    matriculas = row.matricula
    # Generar matrículas aleatorias para el alumno actual
    for i in range(matriculas):
        # Generar una matrícula aleatoria seleccionando un ID de curso con probabilidades definidas por 'prob'
        mat = [id_curso, np.random.choice(cursos.index, p=prob)]
        todas_matriculas.append(mat)

# Crear un DataFrame 'matriculas' a partir de la lista 'todas_matriculas' con las columnas 'id_alumno' e 'id_curso'
matriculas = pd.DataFrame(todas_matriculas, columns=['id_alumno', 'id_curso'])

# Mostrar las primeras filas del DataFrame 'matriculas'
matriculas.head()
```

* .groupby()
* .count()

```python
# Agrupa el DataFrame 'matriculas' por 'id_curso' y cuenta las matrículas para cada curso
matriculas_por_curso = matriculas.groupby('id_curso').count()
matriculas_por_curso
```

* .join()

```python
# Une el DataFrame 'matriculas_por_curso' con la columna 'nombre_curso' del DataFrame 'cursos'
matriculas_por_curso = matriculas_por_curso.join(cursos['nombre_curso'])
matriculas_por_curso
```

* raname()

```python
# Selecciona las columnas 'nombre_curso' e 'id_alumno' y las renombra a 'Total'
matriculas_por_curso = matriculas_por_curso[['nombre_curso', 'id_alumno']].rename(columns={'id_alumno': 'Total'})
matriculas_por_curso
```

* .sort_values()

```python
# Ordena el DataFrame 'matriculas_por_curso' en orden descendente basado en la columna 'Total'
matriculas_por_curso = matriculas_por_curso.sort_values('Total', ascending=False)
matriculas_por_curso
```
